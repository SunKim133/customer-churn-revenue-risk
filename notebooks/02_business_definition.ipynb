# 02. Business Definition

## Churn Definition

In the source dataset, churn indicates whether a customer discontinued their service within the most recent billing period.

For the purposes of this analysis, churn is treated as a binary outcome representing the loss of future recurring revenue.

This framing aligns the dataset definition with how subscription-based businesses evaluate short-term retention risk and revenue stability.

## Time Horizon Assumption

Revenue impact is evaluated on a monthly basis.

`MonthlyCharges` is used as a proxy for short-term revenue at risk, assuming churned customers would otherwise continue paying at a similar rate in the near term.

## Revenue at Risk Definition

Revenue at risk is defined as the expected monthly revenue loss associated with customer churn.
It is calculated as:

Expected Revenue Loss = Probability of Churn Ã— Monthly Charges

This approach prioritizes customers based on financial impact rather than churn likelihood alone.
